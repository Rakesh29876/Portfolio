# 6.) Correlation between UnemploymentRate and AverageSalesPrice.

# 7.) Group data by year and find average HomeownershipRate.

```python
df['Year'] = df.index.year
df.groupby('Year')['HomeownershipRate'].mean()
```

# 8.)Month-over-month % change in HomePriceIndex.

```python
df['HomePriceIndex'].pct_change().head()
```

# 9.)Plot a 12-month rolling average for AverageSalesPrice.

```python
df['AverageSalesPrice'].rolling(window=12).mean().plot(title='12-Month Rolling Average Sales Price', figsize=(10,5))
```

# 10.)Best month to sell: Low DOM, high price.

```python
df['Month'] = df.index.month
best_months = df.groupby('Month')[['MedianDaysonMarket', 'MedianListingPriceperSquareFeet']].mean()
best_months.sort_values('MedianListingPriceperSquareFeet', ascending=False).head()
```

# 11.) Monthly Median `AverageSalesPrice_NewHousesSold` (across years)

```python
monthly_median = df.groupby('Month')['AverageSalesPrice_NewHousesSold'].median()
monthly_median.plot(kind='bar', title='Monthly Median of AvgSalesPrice_NewHousesSold', figsize=(10,5), color='red')
```

# 12.)Check region-wise average `UnemploymentRate`

```python
df['Region'].value_counts()
df.groupby('Region')['UnemploymentRate'].mean()
```

# 13.)Supply-Demand Ratio average

```python
df['SupplyDemandRatio'] = df['MonthlySupplyofNewHouses'] / df['NewOneFamilyHousesSold']
df['SupplyDemandRatio'].mean()

```

# 14.)Histogram visualization

```python
cols = ['HomeownershipRate', 'UnemploymentRate']
df[cols].hist(bins=20, figsize=(10,4), layout=(1,2), color='coral')
plt.suptitle("Distributions of Homeownership and Unemployment Rates")
plt.tight_layout()
```

# 15.)Scatter plot of HomePriceIndex vs MedianSalesPriceofHousesSold

```python
df.plot.scatter(x='HomePriceIndex', y='MedianSalesPriceofHousesSold', alpha=0.6, figsize=(8,5), color='green')
```

# 16.)Yearly Growth in AverageSalesPrice

```python
yearly_avg = df.groupby('Year')['AverageSalesPrice'].mean()
yearly_growth = yearly_avg.diff()
yearly_growth.plot(marker='o', title='Yearly Growth in Average Sales Price', figsize=(8,5))
```

# 17.)  Cumulative sum of NewOneFamilyHousesSold

```python
df['CumulativeSold'] = df['NewOneFamilyHousesSold'].cumsum()
df['CumulativeSold'].plot(title='Cumulative One-Family Houses Sold', figsize=(10,5))
```

# 18.) Compare New vs Overall Average Sales Price

```python
df[['AverageSalesPrice_NewHousesSold', 'AverageSalesPrice']].plot(figsize=(12,5), title='New vs Overall Sales Price')

```
